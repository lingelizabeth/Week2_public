```
signal = np.array(sample2[str('NeuLog experiment')])
signal = signal.astype(float)
signal = np.nan_to_num(signal)
x = np.linspace(0, 10, 501)
#print(x)
y = []
print(np.fft.rfft(signal, n=1000).shape)
for c, i in enumerate(np.fft.rfft(signal, n=1000)): #enumerate gives a count and what's actually within rfft
  y.append(np.abs(i))

#print(y)
plt.figure(figsize=(20, 7))
plt.plot(x, y)
plt.show()
```
